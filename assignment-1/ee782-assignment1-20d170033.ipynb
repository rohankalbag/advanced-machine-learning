```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import torch
```

```python
def visualize_stocks(filename, column_names=['open', 'close', 'high', 'low']):
    df = pd.read_csv(filename, header=0, names=["datetime", "open", "high", "low", "close", "volume"])
    df['datetime'] = pd.to_datetime(df['datetime'], format='%Y-%m-%d %H:%M:%S')
    for i in column_names:
        plt.plot(df['datetime'], df[i], label=i)
    plt.title(filename.split('/')[1].split('_')[0])
    plt.xlabel("Datetime in Years")
    plt.ylabel("Value")
    plt.legend()
    plt.show()

visualize_stocks('sp500_tickers_A-D_1min_1pppix/A_1min.txt')
```
